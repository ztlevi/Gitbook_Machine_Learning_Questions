```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

### Helper Functions

```python
def plot_hist(x, bins):
    plt.hist(x, bins, histtype='bar', density=False, rwidth=0.1)
    plt.axis([bins[0] - 1, bins[-1] + 1, 0, x.shape[0]])
    plt.show()
    
def visualize_data(X, Y):
    colors = itertools.cycle(["r", "b"])
    plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color=next(colors))
    plt.scatter(X[Y == 0, 0], X[Y == 0, 1], color=next(colors))
    plt.show()
```

### Import Data

```python
n = 2000
featureSize = 2
minV = -0.02
maxV = 0.02
X = np.random.uniform(minV, maxV, (n, featureSize))

# here define a ellipse data
Y = np.zeros((n,))

boundary = np.sqrt(X[:,0] * X[:,0] / (2.0 * 2.0) + X[:,1]*X[:,1])
Y[boundary <= 0.01] = 1
```

```python
visualize_data(X, Y)
```

### Data Normalization

```python
X_max = np.max(X, axis=0)
X_min = np.min(X, axis=0)
X = (X - X_min) / (X_max - X_min) - 0.5
```

```python
visualize_data(X, Y)
```

### Random Split

```python
train_ratio = 0.8
val_ratio = 0.1
random_index = np.arange(0, n)
np.random.shuffle(random_index)
idx0 = np.round(train_ratio * n).astype(np.int32)
idx1 = np.round(train_ratio * n + val_ratio * n).astype(np.int32)
                       
X_train = X[random_index[:idx0], :]
Y_train = Y[random_index[:idx0]]
X_val = X[random_index[idx0:idx1], :]
Y_val = Y[random_index[idx0:idx1]]
X_test = X[random_index[idx1:], :]
Y_test = Y[random_index[idx1:]]
```

```python
visualize_data(X_train, Y_train)
visualize_data(X_val, Y_val)
visualize_data(X_test, Y_test)
```
