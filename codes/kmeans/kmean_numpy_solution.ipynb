# K Mean  Using Numpy

```python
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
import numpy as np
import copy

# online example to generate random sample
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.80, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.show()
```

```python
#initilization
k = 4
n_samples = 300
n_feature = 2
idxes = np.random.choice(np.arange(n_samples), k)
centroids = X[idxes]
old_centroids = centroids
distances = np.zeros((n_samples, k), dtype=np.float32)

error = 100

threshold = 0.01

fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], s=7, c="b")
ax.scatter(centroids[:, 0], centroids[:, 1], marker="*", s=200, c="#050505")
plt.show()
```

```python
#update k cluster
while error > 0:
    for i in range(k):
        distances[:, i] = np.sum((X - centroids[i])**2, axis=1)
    belong = np.argmin(distances, axis=1)
    old_centroids = copy.deepcopy(centroids)
    for i in range(k):
        centroids[i] = np.mean(X[belong==i], axis=0)

    error = np.sum((centroids-old_centroids) ** 2)
    error = np.asscalar(error)
    print("error =", error)
```

```python
fig, ax = plt.subplots()
colors = ["r", "g", "b", "y", "c", "m"]
for i in range(k):
    points = X[belong == i, :]
    ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(centroids[:, 0], centroids[:, 1], marker="*", s=200, c="#050505")
plt.show()
```
